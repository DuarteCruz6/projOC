## d) Evaluate the obtained speedup, when compared to the base setup, considered in Section 2.1.

SpeedUp = T old / T new = ((1 / Clock Rate old) x #Cycles old) / ((1 / Clock Rate new) x #Cycles new) = (Clock Rate new / Clock Rate old) x 2.52459

If we assume the clock rate is equal, then SpeedUp = 2.52459

But, because the second processor has data forwarding, it's safe to assume that the clock rate will be lower (the extra logic lengthens the critical path).

That said, the SpeedUp will probably be lower than 2.52459
