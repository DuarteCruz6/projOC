### 2- What is the cache capacity?

Based on the graph, we can see that the time for read+write with arrays of size <= 64KB is small and constant. From 128KB onwards, the time gets really big. Hence, we can assume that the cache capacity for the computer used is 64KB.

### 3- What is the size of each cache block?

Based on the graph, we can see that for arrays of size <=64KB, the time is constant for stride <=64. From stride = 128 onwards, the time flunctuates a lot. Therefore, we can assume that each cache block has a capacity of 64B.

### 4- What is the L1 cache miss penalty time?

To measure the penalty, we need to use the same stride value for an array whose size fits in our cache and for an array whose size does not fit in our cache.

Therefore, we can use stride = 128 and arrays of size 64KB and 128KB each.

L1_miss_penalty_time= array128_stride128 - array64_stride128
                    = 1000 - 320 = 680ns

So, the L1 cache miss penalty time is 680 nanoseconds.
