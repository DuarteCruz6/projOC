### 2- What is the cache capacity?

Based on the graph, we can see that the time for read+write with arrays of size <= 64KB is small and constant. From 64KB to 128KB, there is a noticeable jump in time and we see that the time gets really big. Hence, we can assume that the cache capacity for the computer used is 64KB.

### 3- What is the size of each cache block?

From the graph, we see that the read+write time increases sharply until stride = 16 in the arrays with a size bigger than the cache size. 

We also know that when the stride equals the cache block size, every access maps to a different block, leading to more cache misses. 

Therefore, the cache block must hold 16 words, which corresponds to 16B.

### 4- What is the L1 cache miss penalty time?

To measure the penalty, we need to use the same stride value for an array whose size fits in our cache and for an array whose size does not fit in our cache.

Therefore, we can use stride = 32 and arrays of size 64KB and 128KB each.

In the array of size 64KB, there is only hits, so the time will be the product of the T_hit with the number of iterations. Therefore, array64_stride32 = N_ITERATIONS * T_hit. Based on the graph, the array64_stride32 is equal to 375 nanoseconds.

In the array of size 128KB, there is only misses, so the time will be the product of the Tmiss with the number of iterations. The Tmiss corresponds to the sum of T_hit with the miss penalty time. Therefore, array128_stride32 = N_ITERATIONS * (T_penalty + T_hit). Based on the graph, the array128_stride32 is equal to 1000 nanoseconds.

array128_stride32 - array64_stride32                                      = 1000 - 375 <=>
N_ITERATIONS * (T_penalty + T_hit) - N_ITERATIONS * T_hit                 = 625        <=>
N_ITERATIONS * T_penalty + N_ITERATIONS * T_hit - N_ITERATIONS * T_hit    = 625        <=>
N_ITERATIONS * T_penalty                                                  = 625        <=>
T_penalty = 625 / N_ITERATIONS

Assuming N_ITERATIONS = 100 * stride, just like in our C program, we can conclude that the L1 cache miss penalty time is 0.195 nanoseconds.
